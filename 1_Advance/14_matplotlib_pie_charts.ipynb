#### **Pie Charts**

```python
#importing lib
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast
#load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill_list : ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df.head()
```

```python
df['job_work_from_home'].value_counts().plot(kind="pie", startangle=90, autopct='%1.1f%%')#important-'autopct'
```

### **Bad example for using pie charts**

```python
df['job_title_short'].value_counts().plot(kind='pie',)
plt.title('Job Titles in the Dataset')
plt.ylabel('')
plt.show()
```

```python
fig, ax = plt.subplots(1,3)
dict_column = {
    'job_work_from_home':'Work from Home',
    'job_no_degree_mention':'Job Degree Req.',
    'job_health_insurance':'Health Insurance Offered'
}
for i, (column, title) in enumerate(dict_column.items()):#column and title are tuples therefore using ()
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.1f%%', labels=['False','True'])
    ax[i].set_title(title)
plt.show()
```
